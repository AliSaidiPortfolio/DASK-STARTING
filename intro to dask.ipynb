```python
import dask 
@dask.delayed #create graph task function 
def inc(x):
    return x+1
```

```python
@dask.delayed
def add(x,y):
    return x+y
```

```python
a=inc(1) #create graphe and dont execute 
a
```

```python
b=inc(2)
```

```python
c=add(a,b)
c
```

```python
c=c.compute() #run all graph task include functions,run all triggers
c
```

```python
from dask.distributed import Client
client=Client()
```

```python

def inc(x):
    return x+1
def add(x,y):
    return x+y
```

```python

a=client.submit(inc,1)
a
```

```python
b = client.submit(inc, 2)
b
```

```python
c = client.submit(add, a, b)
c
```

```python
c.result()
```

```python
#every processus  wait for other to done to start execution
```
